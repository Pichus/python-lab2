##### Скільки файлів з кожним розширенням?


Відповідну інформацію вивести на екран у табличному вигляді. Рядки таблиці містять розширення, кількість файлів та ознаку (C,W,O відповідно для файлів з текстом програм (розширення з списку <code>CODE_LIST</code>), інших з білого списку, та тих, що не потрапили в білий список). Інформація подається за спаданням кількості файлів.


Спочатку треба підсумувати інформацію по кожному розширенню. Допоможе метод <code>groupby</code> <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html#pandas.DataFrame.groupby" target=_blank rel="noopener noreferrer">документація тут</a>.

Групувати слід за розширенням (параметр <code>by</code>) і для кожної групи обчислити суму кількостей файлів (метод <code>sum</code>).

Щоб не сумувати роки, місяці тощо, групування слід застосувати до вибірки потрібних стовпчиків. Маємо вибрати стовпчик, за яким групуємо, і стовпчик, який сумуємо.

<details>
    <summary>Підказка</summary>
    Щоб вибрати потрібні стовпчики, можна використати операцію індексації по сукупності стовпчиків, наприклад, <code>dataFrame[['columnA','columnB']]</code>
</details>

Результат присвоїти змінній <code>task1</code>.


```python
task1 = data[["ext", "cnt"]].groupby(by="ext").sum()
task1
```

Відсортувати за спаданням кількості файлів. Допоможе метод <code>sort_values</code> (<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html" target=_blank rel="noopener noreferrer">документація тут</a>).

Не забудьте вказати, що сортування відбувається за спаданням та "на місці".

<details>
    <summary>Підказки</summary>
    <ol>
        <li>параметр <code>by</code> може задавати по яких стовпчиках сортувати</li>
        <li>параметр <code>ascending</code> задає спадання-зростання</li>
        <li>параметр <code>inplace</code> задає чи виконується перетворення на місці або створюється копія</li>
    </ol>
</details>


```python
task1.sort_values(by="cnt", inplace=True, ascending=False)
task1.head(3)
```

За умовою ще треба додати стовпчик з ознакою категорії розширення файлу W,C,O (див. умову).
У лабораторній роботі №1 була написана функція <code>kind</code>, що за розширенням повертає рядок з шуканою ознакою. Використаємо її.


```python
def kind(extension):
    if extension in CODE_LIST:
        return "C"
    if extension in WHITE_LIST:
        return "W"
    return "O"
```

Додамо до таблиці стовпчик з ознаками розширень файлів.


Зараз розширення є індексом. За допомогою методу метод <code>reset_index</code> (<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html" target=_blank rel="noopener noreferrer">документація тут</a>)
зробимо його стовпчиком таблиці.

Не забувайте виконувати перетворення "на місці".


```python
task1.reset_index(inplace=True)
task1.head(3)
```

Тепер вже додаємо стовпчик з ознаками.

<details>
    <summary>Підказки</summary>
    <ol>
        <li>Щоб додати до таблиці <code>t</code> стовпчик <code>name</code>, можна виконати присвоювання виразу <code>t['name']</code></li>
        <li>Присвоювати можна скаляр (не наш випадок) або інший стовпчик.</li>
        <li>Щоб отримати стовпчик з ознаками, можна застосувати метод <code>apply</code>  (<a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html#pandas.Series.apply" target=_blank rel="noopener noreferrer">документація тут</a>) до стовпчика з розширеннями.</li>
    </ol>

```python
task1["kind"] = task1["ext"].apply(kind)
task1.head(3)
```

Групувати слід за роком та місяцем (параметр <code>by</code> має бути списком). Оскільки далі будемо обчислювати статистики для розміру та кількості рядків, то перед групуванням з таблиці слід вибрати стовпчики, за якими групуємо і за якими обчислюємо статистики.


```python
task2 = selection[["year", "mon", "cnt", "size", "lines"]].groupby(by=["year", "mon"])
```

Далі необхідно знайти кількість файлів за кожен місяць, їх сумарний та середній розмір, а також сумарну та середню кількість рядків.

У результаті групування отримався об'єкт <code>DataFrameGroupBy</code>
(<a href="https://pandas.pydata.org/docs/reference/groupby.html" target=_blank rel="noopener noreferrer">документація тут</a>).

До нього застосовний метод <code>agg</code>, що обчислює статистики (агрегатні характеристики) (<a href="https://pandas.pydata.org/docs/reference/api/pandas.api.typing.DataFrameGroupBy.agg.html#pandas.api.typing.DataFrameGroupBy.agg" target=_blank rel="noopener noreferrer">документація тут</a>).

Для стовпчика <code>cnt</code> будемо шукати суму значень (<code>sum</code>), для стовпчика <code>size</code> будемо шукати суму значень та середнє значення (<code>mean</code>), для стовпчика <code>lines</code> будемо шукати суму значень та середнє значення.

Першим параметром методу <code>agg</code> зробимо словник, що задає які статистики для яких стовпчиків слід обчислювати. Ключами словника мають бути імена стовпчиків (рядки), а значеннями списки статистик або окремі статистики (статистики також задаємо рядками-іменами)).

Результат агрегування зробимо значенням змінної <code>res2</code>.


```python
res2 = task2.agg({"cnt": "sum", "size": ["sum", "mean"], "lines": ["sum", "mean"]})
res2.head()
```

Оскільки відбувалось групування по кільком стовпчикам та для окремих стовпчиків обчислювалось кілька характеристик, то отримали складені індекси (multindex) і рядків, і стовпчиків.

Поточний індекс стовпчиків.


```python
res2.columns
```

Перейдемо до простого індексу стовпчиків.


```python
res2.columns = res2.columns.to_flat_index()
res2.columns
```

```python
res2.head(3)
```

Змінимо назви стовпчиків.


```python
res2.columns = [
    "Кількість",
    "Сумарний розмір",
    "Середній розмір",
    "Кількість рядків",
    "Середня кількість рядків",
]
res2.head(3)
```

Виразимо сумарний розмір у kb.

Для цього слід поділити значення на 1024, округлити. Додатково перетворимо на цілий тип (метод <code>astype</code> дозволяє виконувати перетворення типів, <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.astype.html" target=_blank rel="noopener noreferrer">документація тут</a>). Також не забудемо змінити підпис стовпчика.


Переходимо до кілобайтів.


```python
res2["Сумарний розмір"] = (res2["Сумарний розмір"] / 1024).round().astype("int64")
```

Змінюємо підпис стовпчиків


```python
res2.rename(
    columns={
        "Сумарний розмір": "Сумарний розмір, kb",
        "Середній розмір": "Середній розмір, b",
    },
    inplace=True,
)
```

```python
res2.head(3)
```

Округлити середній розмір та середню кількість рядків, перейти до цілих значень.


```python
res2.info()
```

Ось тільки тепер замінимо відсутні значення кількості рядків нулями!

Якщо б це було зроблено раніше, то середня кількість рядків була б обрахована неправильно ( на неї вплинули б нульові значення замість відсутніх).

<details>
    <summary>Підказка</summary>
    Використати метод <code>fillna</code> (<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html" target=_blank rel="noopener noreferrer">документація тут</a>). Оскільки відсутні дані тільки в одному стовпчику, то все просто. Тільки не забудьте, що треба перетворювати наявний датасет, а не створювати новий.
</details>


```python
res2.fillna({"Кількість рядків": 0}, inplace=True)
res2.info()
```

Тепер можна перетворювати типи стовпчиків.


```python
fname = 'Середній розмір, b'
res2[fname] = res2[fname].round().astype("int64")

fname = 'Середня кількість рядків'
res2[fname] = res2[fname].round().astype("int64")
```

```python
res2
```

Відсортуємо в порядку, оберненому до хронологічного.

У даному випадку слід сортувати індекс.
Допоможе метод <code>sort_index</code> (<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_index.html" target=_blank rel="noopener noreferrer">документація тут</a>).


```python
res2.sort_index(ascending=[False, False], inplace=True)
res2
```
